```python
# reading previous stored links
all_urlsdf=pd.read_csv('links.csv')
all_urls= all_urlsdf.iloc[:, 1].tolist()
```

### Webscrapping details of car from gathered links

#### dont run this

```python
links_sum=0
all_names=[]# stores names of car
all_price=[]# stores price of car
all_rating=[]# stores rating of car
all_car_info=[]# list stores all other details of car
city=[]

for idx,url in tq.tqdm(enumerate(all_urls)):         # iterating through each url to get details of each car details
    car_page=requests.get(url)
    soup_car=bs.BeautifulSoup(car_page.content,'html.parser') # applying beautiful soup  html parser
    
    
    details=soup_car.find_all('div',class_='media-body _3fGRT') # finding all class media-body_3fGRT for getting multiple details of car
    price=soup_car.find('h4',class_='_3NnL0') # getting price
    name=soup_car.find('p',class_='_342La') # getting name of car
    rating=soup_car.find('p',class_='_3bipb') # getting rating of car
    
    # checking if these fields are empty
    if None in (price,name,rating):
        #print('i')
        continue
        
    # storing all gathered fields
    all_names.append(name.text)
    all_price.append(price.text)
    all_rating.append(rating.text)
    city.append(all_urlsdf['city'][idx])
    car_info=[]
    for info in details:
        car_info.append(info.text)
    all_car_info.append(car_info)
    
    links_sum+=1
    print(links_sum) # printing number of successful link fetches
            
#print(all_car_info)
#print(all_names)
#print(all_rating)
#print(all_price)
```

```python
# Substituting nan in all_car_info where values are missing
# since some values are missing or not given ,the len of every list in all_car_info is less than 8
# this code checks if length of every list in list all_car_info is 8 and if not substitutes nan values in desired position
# making len of list to be 8
for i in range(len(all_car_info)): # iterates for every field in all_car_info
    #print(i[4])
    print(all_car_info[i])
    if len(all_car_info[i])<8: #if len of all_car_info < 8 the below code gets executed
        
        if(((all_car_info[i][4]!='TransmissionMANUAL') or (all_car_info[i][4]!= 'TransmissionAUTOMATIC')) 
            and (len(all_car_info[i])==6)):
            
            all_car_info[i].insert(4,np.NaN) # substituting nan
            all_car_info[i].insert(6,np.NaN)
       
        elif(all_car_info[i][4]=='TransmissionMANUAL') or (all_car_info[i][4]== 'TransmissionAUTOMATIC'):
            all_car_info[i].insert(6,np.NaN)
        
        else:
            all_car_info[i].insert(4,np.NaN)
    print(i)
    
```

```python
dic={'Name':all_names,'Price':all_price,'Rating':all_rating,'city':city} # creating dict object of every field gathered
```

```python
data=pd.DataFrame(dic) # creating data frame for further use
d=pd.DataFrame(all_car_info,columns=['Kilometers','Year_of_Purchase','Owner','Fuel_Type',
                                     'Transmission','RTO','Insurance','Insurance_Type'])
```

```python
dataset=pd.concat([data,d],axis=1) # concating  'Price','Rating','city':city ,'names' with 'all_car_info'
```
