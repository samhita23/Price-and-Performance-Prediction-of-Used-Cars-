### Web scraping code

#### dont run the below code cell

```python
# Scraping links of every car
total_urls=[]
total_links=0
city=[]

city_code=[3686,5732,4709,777,2423,2378,290,1692,2130,769] # TEN CITY codes
for code in tqdm(city_code): # loop iterates for every city
    for i in tqdm(range(1,601)):# loop iterates for 600 pages in every city
        try:
            page=requests.get('https://www.cars24.com/buy-used-car?sort=P&storeCityId='+str(code)+'&page='+str(i)) # url of site
            soup=BeautifulSoup(page.content,'html.parser')
            link=soup.find_all('a',class_='_20d39')# class contining url of every car in given page
            for i in link:
                href=i.get('href') # getting links 
                total_urls.append(href) # storing it in lisy
                total_links=total_links+1 # counting number of links obtained
                city.append(code) # append city code for every link
        except:
            break
      #print(total_links)
```

```python
len(total_urls)
```

```python
# storing links to data frame for further use
links=pd.DataFrame(total_urls)
links['city']=city
```
